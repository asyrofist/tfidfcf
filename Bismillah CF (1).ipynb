```python
from nltk.tokenize import TweetTokenizer
import numpy as np
def tokenize(text):
    
    TT = TweetTokenizer()
    hasil = TT.tokenize(text)
    filtered = []
    
    for w in hasil:
        filtered.append(w)
        
    return filtered

st = lambda x: tokenize(x)
```

```python
value = []
texts = []
fgh = []
    
for i in range(len(data)):
    # Tokenize
    tweet = pd.DataFrame(data.text.apply(st))
    bvc = tweet.loc[i]
    # to array
    arr = np.array(bvc)
    bvb = np.vstack(bvc)
    # word in dataframe
    dfx = pd.DataFrame(data=bvb.flatten())
    dfx['text'] = dfx
    dfc = pd.DataFrame(dfx['text'])
    dfc

    # count word in one doc
    v = dfc.text.apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0)
    vc = pd.DataFrame(v)
    vc.columns = ['value']

    # array for each value
    for b in vc['value']:
        TF = b / len(vc)
        hv = []
        hv.append(TF)
        fgh.append(hv)

    # assign TF value for word 
    value.append(fgh)
    value
    texts.append(bvc)
    texts
    
vcx = pd.DataFrame(texts)
vcx
```

```python
processed_token_set = []
for i in range(3):
    processed_tokens = []
    for token in raw_token_set[i]:
        processed_tokens.append(token.lower())
    processed_token_set.append(processed_tokens)
```

```python
terms = []
for i in range(3):
    terms.append(sorted(list(set(processed_token_set[i]))))
```

```python
collection_terms = []
for i in range(3):
    collection_terms += terms[i]
collection_terms = sorted(list(set(collection_terms)))
```

```python
raw_tf = []
weighted_tf = []
for i in range(3):
    raw_frequency = []
    weighted_frequency = []
    for term in collection_terms:
        raw_count = processed_token_set[i].count(term)
        if(raw_count != 0):
            weighted_count = 1 + log(raw_count, 10)
        else:
            weighted_count = 0
        raw_frequency.append(raw_count)
        weighted_frequency.append(weighted_count)
    raw_tf.append(raw_frequency)
    weighted_tf.append(weighted_frequency)
```
